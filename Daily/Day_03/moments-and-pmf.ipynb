```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

y = np.arange(0, 10, 0.001)
plt.plot(y, binom.pmf(y, 10, 0.5))
```

```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

mu = 500
x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(x, mu))
```
