```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(1000)
y = np.random.randn(1000)

plt.scatter(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

values = np.random.normal(10000, 1000, 10000)
plt.hist(values, 50)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

values = np.random.rand(100) * 100 - 40
high = np.random.rand(10) * 50 + 100
low = np.random.rand(10) * -50 - 100

data = np.concatenate((values, high, low))
plt.boxplot(data)
plt.show()
```
