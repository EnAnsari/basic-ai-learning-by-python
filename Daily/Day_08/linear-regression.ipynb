```python
import numpy as np
import matplotlib.pyplot as plt

pageSpeeds = np.random.normal(3, 1, 1000)
purchasesAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3

plt.scatter(pageSpeeds, purchasesAmount)
plt.show()
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchasesAmount)
print(slope, intercept, r_value, p_value, std_err)
```

```python
def predict(x):
    return slope * x + intercept

x = 0
print(predict(x))
```

```python
fitLine = predict(pageSpeeds)

plt.scatter(pageSpeeds, purchasesAmount)
plt.plot(pageSpeeds, fitLine, c='r')
plt.show()
```
